<h2 align="center">Image Augmentation</h2>

<h3 align="center">Problem 1</h3>

```python
def get_coordinate(d1,d2):
    #return the coordinate matrix used in translation
    return np.mgrid[0:d1,0:d2].reshape(2,d1*d2)

def apply_to_data(d1,d2,X,new_coords):
    #apply the translation to dataset
    x, y = new_coords.reshape((2, d1, d2), order='F')
    translated = np.take(X, x+d1*y, mode='wrap').reshape((d1, d2))
    return translated

def translate(X, A, B): 
    """Returns each image translated by a random amount (a,b), where
    a~Uniform(-A,A) and b~Uniform(-B,B). The resulting image should be cropped
    to be of size (d1,d2). Note that this translation will leave a border on
    two sides of the image. Fill the empty border with the parts that were cropped
    off the opposite sides.
        Parameters:
            X (d1,d2): d1 x d2 array 
            A (float): horizontal translation parameter
            B (float): vertical translation parameter
        Returns:
            translated_image (d1,d2): array of translated images
    """
    #get shape
    d1,d2 = X.shape
    #get the tranaslation amounts
    a = np.random.uniform(-A,A)
    b = np.random.uniform(-B,B)
    #get the coordinate matrix
    coords = get_coordinate(d1,d2)
    #now do the transformation on the coordinates
    new_coords = coords+np.array([[a],[b]])
    new_coords = new_coords.astype(int)
    #now apply to the image
    translated = apply_to_data(d1,d2,X,new_coords)
    return translated
    
def rotate(X, theta): 
    """ Returns an image rotated by a random amount t~Uniform(-theta,theta).
    The resulting image should be cropped to be the same size as the original,
    and any blank parts should be filled with one of the parts cropped off the
    other side.
        Parameters:
            X (d1,d2): d1 x d2 array
            theta (float): largest acceptable rotation angle
        Returns:
            rotated_image (d1,d2): array of rotated image
    """
    #get shape
    d1,d2 = X.shape
    #get the coordinate matrix
    coords = get_coordinate(d1,d2)
    #create translation matrix
    trans = np.array([[np.cos(theta), -np.sin(theta)],[np.sin(theta), np.cos(theta)]])
    #translate coordinate matrix
    new_coords = trans@coords
    new_coords = new_coords.astype(int)
    translated = apply_to_data(d1,d2,X,new_coords)
    return translated
    
def skew(X, A): 
    """ Returns an image with the linear transformation [[1,a],[0,1]] applied,
    where a~Uniform(-A,A). Crop the parts that go outside the image boundaries
    and fill missing areas with the appropriate cropped piece.
        Parameters:
            X (d1,d2): d1 x d2 array
            A (float): skew parameter
        Returns:
            skewed_image (d1,d2): array of skewed image
    """
    #get shape
    d1,d2 = X.shape
    a = np.random.uniform(-A,A)
    #get the coordinate matrix
    coords = get_coordinate(d1,d2)
    #create translation matrix
    trans = np.array([[1, a],[0,1]])
    #translate coordinate matrix
    new_coords = trans@coords
    new_coords = new_coords.astype(int)
    translated = apply_to_data(d1,d2,X,new_coords)
    return translated    

def flip_horizontal(X):
    """Flip horizontal. Returns a horizontally flipped version of each image.
        Parameters:
            X (d1,d2): d1 x d2 image array 
        Returns:
            flipped_image (d1,d2): array of flipped image
    """
    d1,d2 = X.shape
    #get the coordinate matrix
    coords = get_coordinate(d1,d2)
    #create translation matrix
    new_coords = coords * np.array([[1],[-1]])
    new_coords = new_coords.astype(int)
    translated = apply_to_data(d1,d2,X,new_coords)
    return translated
    
def gauss_noise(X, sigma2): 
    """Adds Gaussian noise with parameter sigma2. For each image draw d1xd2 random
    noise values from N(0,sigma2) and add those to the original image.
        Parameters:
            X (d1,d2): d1 x d2 array 
        Returns:
            noisy_image: array of noisy image
    """
    #get shape
    d1,d2 = X.shape
    #get the tranaslation amounts
    gauss = np.random.normal(0,np.sqrt(sigma2),size=(d1,d2))
    #now apply to the image
    translated = X+gauss
    return translated
    
```

```python
#load the lion data
lion = imread('lion.png')
A = 100
A_1 = 0.5
B = 50
theta=np.pi/3
sigma2 = 100
#translate
translate_lion = translate(lion,A,B)
rotate_lion = rotate(lion,theta)
skew_lion = skew(lion,A_1)
flip_lion = flip_horizontal(lion)
gauss_lion = gauss_noise(lion,sigma2)
#show the transformed plots
fig,ax = plt.subplots(2,3, figsize=(15,12))
ax[0][0].imshow(lion, cmap='gray')
ax[0][0].set_title("Original Image")
ax[0][0].axis('off')
ax[0][1].imshow(translate_lion, cmap='gray')
ax[0][1].set_title("Translated Image: A = {}, B = {}".format(A,B))
ax[0][1].axis('off')
ax[0][2].imshow(rotate_lion, cmap='gray')
ax[0][2].set_title("Rotated Image: theta = {}".format(theta))
ax[0][2].axis('off')
ax[1][0].imshow(skew_lion, cmap='gray')
ax[1][0].set_title("Skewed Image: A = {}".format(A_1))
ax[1][0].axis('off')
ax[1][1].imshow(flip_lion, cmap='gray')
ax[1][1].set_title("Flipped Image")
ax[1][1].axis('off')
ax[1][2].imshow(gauss_lion, cmap='gray')
ax[1][2].set_title("Guassian Image: Sigma2 = {}".format(sigma2))
ax[1][2].axis('off')
plt.show()
```

<h3 align="center"> Problem 2 </h3>
The file music.npy contains the audio time series data from 10 second clips of 150 different songs, with style.npy describing the associated style of ballroom dance. The styles included are Chacha, Foxtrot, Jive, Samba, Rumba, and Waltz. Use train_test_split
from sklearn.model_selection with test_size=.5 to create train and test sets. Create two training sets by augmenting this original training set. Each new augmented
training set will include the original data and the augmented data. For the first, add ambient
noise from the file restaurant-ambience.wav. For the second, use time_stretch.

```python
def compute_plp(data, length):
    data_plp = [librosa.beat.plp(data[i]) for i in range(length)]
    #now convert to numpy array
    data_plp = np.array(data_plp)
    return data_plp

def load_and_addnoise(train_len=120, stretch_rate=0.5):
    #load the data
    y = np.load('styles.npy')
    X = np.load('music.npy')
    m,music_length = X.shape
    restaurant, sample_rate = librosa.load('restaurant-ambience.wav', sr=22050)
    #fancy slicing to get the random samples from restaurant array
    random_indeces = list(np.random.choice(np.arange(0,len(restaurant)-music_length,1,dtype=int),120))
    restaurant_selections = np.array([restaurant[i:i+music_length] for i in random_indeces])
    #get the training data
    x_train = X[0:train_len]
    x_test = X[train_len:]
    y_train = y[0:train_len] 
    y_test = y[train_len:]
    #add the restaurant ambient noise
    restaurant_data = x_train + restaurant_selections
    #now stretch the data
    stretch_data = [librosa.effects.time_stretch(x_train[i,:],
                    rate=stretch_rate) for i in range(train_len)]
    stretch_data = np.stack(stretch_data)
    #create test_data
    x_test = compute_plp(x_test, m-train_len)
    #get the size
    test_r, test_col = x_test.shape
    #compute the plps and name data
    x_train = compute_plp(x_train, train_len)
    rest_train = compute_plp(restaurant_data,train_len)
    stretch_train = compute_plp(stretch_data,train_len)
    #resize stretch_train to fit with other data
    stretch_r, stretch_col = stretch_train.shape
    stretch_train = stretch_train[:,stretch_col-test_col:]
    #now concatenate the data
    rest_train = np.vstack([x_train, rest_train])
    stretch_train = np.vstack([x_train, stretch_train])
    #return the data
    return [x_train, rest_train, stretch_train], y_train, x_test, y_test
    
def random_forest_test_score(x_train, x_test, y_train, y_test, num_train=10):
    #initialize random forest classifier
    rf = RandomForestClassifier()
    scores = list()
    for run in range(num_train):
        #train the random forest, plp was added in the previous step
        rf.fit(x_train, y_train)
        y_predict = rf.predict(x_test)
        scores.append(accuracy_score(y_test, y_predict))
    #find the mean and std    
    mean = np.mean(np.array(scores))
    std = np.std(np.array(scores))
    return mean, std 
        
```

```python
train_set_list, y_train, x_test, y_test = load_and_addnoise()
```

<h3 align="center"> Problem 4 </h3>

Write a function that uses the synthetic minority oversampling technique to augment an imbalanced data set.
Your function should have the following characteristics:

Accept a matrix of minority class samples, the number of samples to generate per original point, and the number of nearest neighbors.

For each original point in the sample, pick one of the $k$ nearest neighbors at random and generate a new point that lies between the two original values. 

Return an array containing the synthetic samples.


```python
def smote(T,N,k):
    """ Generate synthetic points using the SMOTE method. Hint! Use KDTree :)
    Parameters:
        T (n,m): minority class samples
        N (int): number of samples to generate from each point
        k (int): number of nearest neighbors
    Returns:
        synthetic ndarray(N*n,m): synthetic minority class samples
    """
    #get the shape
    n,m = T.shape
    #initialize new matrix
    synth = np.zeros((N*n,m))
    #generate the new points
    itera = 0
    #create the tree
    tree = KDTree(T)
    for i in range(n):
        value = T[i:i+1]
        #find the nearest neighbors
        dist, neighbors = tree.query(value,k=k+1)
        for j in range(N):
            #generate the random neighbor
            neighbor = T[np.random.choice(neighbors.ravel()[1:]),:]
            #generate the random point
            point = np.random.uniform(value,neighbor)
            #add the point to the matrix
            synth[itera,:] = point
            #update the iteration counter
            itera += 1
    return synth
```

<h3 align="center"> Problem 5 </h3>

The dataset found in creditcard.npy contains information about credit card purchases made over a two day period. 
Of the approximately 285,000 observations, 492 are fraudulent purchases. 
The last column indicates if the purchase was valid (0) or fraudulent (1). 

Do the following steps 10 times:

 Create a training and test set from the data using train_test_split from sklearn.
model_selection with test_size=.7.

 Use smote with N = 500 and k = 2 to augment the training set.

 Create two Gaussian Nave Bayes classifiers (from sklearn.naive_bayes.GaussianNB),
one which wil be trained on only the original data and the other on the SMOTE augmented
data and the original data.

 Fit each classifier and find the recall and accuracy of each model.

Print the mean recall and mean accuracy of each model and and describe the findings.

```python
#load the data
credit_data = np.load('creditcard.npy')
X_data = credit_data[:,:-1]
y_data = credit_data[:,-1]
#now get all the data that has a fraudulent purchase
false_mask = y_data == 1
fraud_data = X_data[false_mask]
#now augment the fraud data with 500 fraudulent observations
print("Creating Synthetic Data...")
synth_data = smote(fraud_data, 500, 2)
synth_y = np.ones(500*492)
#now create the new datasets
synth_total = np.vstack([X_data,synth_data])
y_synth = np.hstack([y_data, synth_y])
#shuffle the augmented data
p = np.random.permutation(len(y_synth))
y_synth = y_synth[p]
synth_total = synth_total[p]
#train on the unaugmented data
print("Training Gaussian Models...")
NB_orig = GaussianNB()
NB_orig.fit(X_data,y_data)
#now get the scores using only the original data
y_orig_pred = NB_orig.predict(X_data)
scores_orig = recall_score(y_data, y_orig_pred)
orig_acc = accuracy_score(y_data, y_orig_pred)
#train on the augmented data
NB = GaussianNB()
NB.fit(synth_total, y_synth)
y_aug_pred = NB.predict(X_data)
scores_aug = recall_score(y_data, y_aug_pred)
aug_acc = accuracy_score(y_data, y_aug_pred)
#print the classification reports
print("Unaugmented Recall Score: {}".format(np.round(scores_orig,4)))
print("Unaugmented Accuracy Score: {}".format(np.round(orig_acc,4)))
print('\n\n')
print("Augmented Recall Score: {}".format(np.round(scores_aug,4)))
print("Augmented Accuracy Score: {}".format(np.round(aug_acc,4)))
```
